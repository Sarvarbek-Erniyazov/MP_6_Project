```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)
```

```python
RAW_DATA_DIR = r"C:\Users\sharg\Desktop\uyga vazifa\unsupervised learning\4-oy\6-project\binance.com\data\raw"
```

```python
data_files = [f for f in os.listdir(RAW_DATA_DIR) if f.endswith(".csv")]
coins_data = {}

for file in data_files:
    coin_name = file.split("_")[0]  # e.g., JUP_historical_data_hourly.csv
    df = pd.read_csv(os.path.join(RAW_DATA_DIR, file), index_col=0, parse_dates=True)
    coins_data[coin_name] = df
    print(f"Loaded {coin_name}: {df.shape} rows x {df.shape[1]} columns")
```

```python
for coin, df in coins_data.items():
    print(f"\n--- {coin} ---")
    display(df.head())
    display(df.describe())
    print(f"Missing values:\n{df.isnull().sum()}")
```

```python
for coin, df in coins_data.items():
    plt.figure()
    plt.plot(df.index, df['Close'], label=f"{coin} Close")
    plt.title(f"{coin} Hourly Close Price")
    plt.xlabel("Datetime")
    plt.ylabel("Price (USDT)")
    plt.legend()
    plt.show()
```

```python
for coin, df in coins_data.items():
    df['Return'] = df['Close'].pct_change()
    mean_return = df['Return'].mean()
    std_return = df['Return'].std()
    print(f"{coin} Hourly Mean Return: {mean_return:.6f}, Std Dev: {std_return:.6f}")
```

```python
daily_data = {}

for coin, df in coins_data.items():
    # Daily OHLC, return, volatility
    daily_ohlc = df['Close'].resample('D').ohlc()
    daily_return = df['Return'].resample('D').sum()
    daily_volatility = df['Return'].resample('D').std()
    
    # Store in dictionary
    daily_data[coin] = {
        'ohlc': daily_ohlc,
        'daily_return': daily_return,
        'daily_volatility': daily_volatility
    }

    # Plot daily open-close after storing data
    plt.figure()
    daily_ohlc[['open', 'close']].plot(title=f"{coin} Daily Open-Close Prices")
    plt.xlabel("Date")
    plt.ylabel("Price (USDT)")
    plt.show()
```

```python
vol_df = pd.DataFrame({coin: daily_data[coin]['daily_volatility'] for coin in coins_data.keys()})
vol_df.plot(title="Daily Volatility Across Coins")
plt.xlabel("Date")
plt.ylabel("Volatility (Std of Returns)")
plt.show()
```

```python
close_prices = pd.DataFrame({coin: df['Close'] for coin, df in coins_data.items()})
corr_matrix = close_prices.corr()
plt.figure()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation of Close Prices Between Coins")
plt.show()
```

```python
rolling_window = 24  # 24 hours = 1 day
for coin, df in coins_data.items():
    df['Rolling_Mean'] = df['Close'].rolling(window=rolling_window).mean()
    df['Rolling_Std'] = df['Close'].rolling(window=rolling_window).std()
    
    plt.figure()
    plt.plot(df.index, df['Close'], label="Close")
    plt.plot(df.index, df['Rolling_Mean'], label="Rolling Mean (24h)")
    plt.fill_between(df.index, df['Rolling_Mean']-df['Rolling_Std'], df['Rolling_Mean']+df['Rolling_Std'],
                     color='gray', alpha=0.3, label="Rolling Std Dev")
    plt.title(f"{coin} Price with 24h Rolling Mean & Std")
    plt.xlabel("Datetime")
    plt.ylabel("Price (USDT)")
    plt.legend()
    plt.show()
```

```python
avg_daily_vol = {coin: daily_data[coin]['daily_volatility'].mean() for coin in coins_data.keys()}
sorted_vol = sorted(avg_daily_vol.items(), key=lambda x: x[1], reverse=True)
print("Coins sorted by average daily volatility (highest first):")
for coin, vol in sorted_vol:
    print(f"{coin}: {vol:.6f}")
```

```python
avg_hourly_return = {coin: df['Return'].mean() for coin, df in coins_data.items()}
sorted_return = sorted(avg_hourly_return.items(), key=lambda x: x[1], reverse=True)
print("\nCoins sorted by average hourly return (highest first):")
for coin, ret in sorted_return:
    print(f"{coin}: {ret:.6f}")
```

```python
# --- Import libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Set plotting style ---
sns.set(style='whitegrid', palette='muted', font_scale=1.2)

# --- Load raw data ---
DATA_DIR = r"C:\Users\sharg\Desktop\uyga vazifa\unsupervised learning\4-oy\6-project\binance.com\data\raw"

coins = ['APE', 'ARB', 'FIL', 'GALA', 'JUP', 'OP', 'SEI', 'TIA']
coins_data = {}

for coin in coins:
    file_path = os.path.join(DATA_DIR, f"{coin}_historical_data_hourly.csv")
    df = pd.read_csv(file_path, parse_dates=['Datetime'], index_col='Datetime')
    # Compute hourly return
    df['Return'] = df['Close'].pct_change()
    coins_data[coin] = df

# --- Summary Statistics ---
summary_stats = {}
for coin, df in coins_data.items():
    summary_stats[coin] = {
        'MeanReturn': df['Return'].mean(),
        'Volatility': df['Return'].std(),
        'AvgVolume': df['Volume'].mean()
    }

summary_df = pd.DataFrame(summary_stats).T
summary_df = summary_df.sort_values('MeanReturn', ascending=False)
summary_df

# --- Plot Cumulative Returns ---
plt.figure(figsize=(12,6))
for coin, df in coins_data.items():
    cum_return = (1 + df['Return']).cumprod()
    plt.plot(cum_return, label=coin)
plt.title("Kripto-valyutalarning soatlik kumulyativ rendimenti")
plt.xlabel("Vaqt")
plt.ylabel("Kumulyativ Rendiment")
plt.legend()
plt.show()

# --- Plot Volatility vs Return (Bubble by Avg Volume) ---
plt.figure(figsize=(10,6))
plt.scatter(
    summary_df['Volatility'],
    summary_df['MeanReturn'],
    s=summary_df['AvgVolume']/50000,
    alpha=0.7
)
for coin in summary_df.index:
    plt.text(summary_df.loc[coin,'Volatility'], summary_df.loc[coin,'MeanReturn'], coin)
plt.title("Xavf vs Rendiment (Balon hajmi ~ Avg Volume)")
plt.xlabel("Soatlik Volatilitet (Std Dev)")
plt.ylabel("Ortacha Soatlik Rendiment")
plt.show()

# --- Correlation Heatmap ---
close_prices = pd.DataFrame({coin: df['Close'] for coin, df in coins_data.items()})
plt.figure(figsize=(8,6))
sns.heatmap(close_prices.corr(), annot=True, cmap='coolwarm')
plt.title("Soatlik Close narxlar korrelyatsiyasi")
plt.show()

# --- Daily Aggregation (OHLC, Return, Volatility) ---
daily_data = {}
for coin, df in coins_data.items():
    daily_ohlc = df['Close'].resample('D').ohlc()
    daily_return = df['Return'].resample('D').sum()
    daily_volatility = df['Return'].resample('D').std()
    daily_data[coin] = {
        'OHLC': daily_ohlc,
        'DailyReturn': daily_return,
        'DailyVolatility': daily_volatility
    }
    # Plot Daily Open-Close
    plt.figure(figsize=(10,4))
    daily_ohlc[['open','close']].plot(title=f"{coin} Daily Open-Close Prices")
    plt.xlabel("Sana")
    plt.ylabel("Narx (USDT)")
    plt.show()
```
