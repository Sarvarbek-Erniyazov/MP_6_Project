## Nima uchun Multi-Class Classification to‘g‘ri tanlov?

Biz oddiy binary model ('Up'/'Down') o‘rniga **multi-class** ('Up'/'Sideways'/'Down') modelni tanladik. Sabablari quyidagicha:

### 1. Ko‘p narx o‘zgarishlari shovqin (noise)
1-soatlik JUP returns histogrami ko‘rsatadi:
- Markaziy cho‘qqi juda katta → bu kichik, ahamiyatsiz narx o‘zgarishlari  
- Binary model buni bashorat qilishi kerak bo‘ladi → imkonsiz va foydasiz  

### 2. Threshold orqali foydali sinflar yaratish
±0.5% threshold bilan:

- **Sideways (To‘xtash)**: -0.5% dan +0.5% gacha, shovqin (~40%)  
- **Up (O‘sish)**: ≥ +0.5% (~30%)  
- **Down (Pasayish)**: ≤ -0.5% (~30%)

Shu bilan model **faqat ahamiyatli narx harakatlarini** o‘rganadi, shovqinni e’tiborsiz qoldiradi.

### 3. Balans va realistik vazifa
Multi-class yondashuv datasetni muvozanatlaydi va modelga **haqiqiy, amaliy target** beradi.  

Binary classification esa faqat **shovqinli va foydasiz signal**larni bashorat qiladi.

### 4. Xulosa
Multi-class classification quyidagilarni beradi:

- Shovqinni e’tiborsiz qoldirish (Sideways sinfi)  
- Ahamiyatli harakatlarni aniqlash (Up/Down)  
- Modelni tushunish va amaliy qiymatini oshirish  

Shuning uchun 1-soatlik JUP return prediction uchun **multi-class optimal yondashuv** hisoblanadi.
